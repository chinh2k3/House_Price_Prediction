# 🏠 House Price Prediction — Preprocessing
> Pipeline: Missing Values → Outliers → Encoding → Feature Engineering → Save

## 1. Library Imports

```python
from pathlib import Path
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style('whitegrid')
plt.rcParams['figure.dpi'] = 100
pd.options.display.max_columns = None
```

## 2. Data Loading

```python
BASE_DIR  = Path.cwd().parent
DATA_PATH = BASE_DIR / "Data" / "raw" / "train.csv"

csv_data = pd.read_csv(DATA_PATH)
print(f"Shape: {csv_data.shape}")
csv_data.head()
```

## 3. Overview

```python
df_raw = csv_data.copy()

print(f"Shape      : {df_raw.shape}")
print(f"Duplicates : {df_raw.duplicated().sum()}")
print(f"\nDtypes:")
print(df_raw.dtypes.value_counts())
```

```python
# Tách X và y, bỏ Id
df_raw = df_raw.drop(['Id'], axis=1)
X = df_raw.drop(['SalePrice'], axis=1)
y = df_raw['SalePrice']
df = X.copy()
print(f"X shape: {df.shape} | y shape: {y.shape}")
```

## 4. Missing Values

```python
# Bảng tổng hợp missing
miss_count   = df.isnull().sum()
miss_percent = miss_count / len(df) * 100
miss_df = pd.DataFrame({
    'Missing Count'  : miss_count,
    'Missing Percent': miss_percent
}).sort_values('Missing Percent', ascending=False)

# Chỉ hiển thị cột có missing
miss_df[miss_df['Missing Count'] > 0]
```

```python
# Visualize missing percent
top_miss = miss_df[miss_df['Missing Percent'] > 0].head(20)
plt.figure(figsize=(10, 5))
top_miss['Missing Percent'].sort_values().plot(kind='barh', color='salmon')
plt.axvline(50, color='red', linestyle='--', label='50% threshold')
plt.xlabel('Missing %')
plt.title('Top columns with missing values')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Xóa cột missing > 50%
thresh  = 50
to_drop = miss_df[miss_df['Missing Percent'] > thresh].index.tolist()
df = df.drop(columns=to_drop)
print(f"Dropped {len(to_drop)} columns: {to_drop}")
print(f"Remaining shape: {df.shape}")
```

```python
# Tách numeric và categorical
num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
cat_cols = df.select_dtypes(include=['object', 'category', 'bool']).columns.tolist()
print(f"Numeric cols   : {len(num_cols)}")
print(f"Categorical cols: {len(cat_cols)}")
```

```python
# Tạo flag was_missing trước khi fill
cols_with_missing = df.columns[df.isnull().sum() > 0]
for col in cols_with_missing:
    df[f"{col}_was_missing"] = df[col].isnull().astype(int)

print(f"Flag columns created: {[c for c in df.columns if c.endswith('_was_missing')]}")
```

```python
# Fill numeric: group by Neighborhood trước, còn lại dùng median
if 'Neighborhood' in df.columns:
    df[num_cols] = df.groupby('Neighborhood')[num_cols].transform(
        lambda x: x.fillna(x.median())
    )

# Fill còn lại bằng median toàn bộ
df[num_cols] = df[num_cols].fillna(df[num_cols].median())

# Fill categorical bằng mode
for col in cat_cols:
    if df[col].isnull().any():
        mode_val = df[col].mode()
        df[col]  = df[col].fillna(mode_val.iloc[0] if len(mode_val) else 'Missing')

print(f"Missing sau khi fill: {df.isnull().sum().sum()}")
assert df.isnull().sum().sum() == 0, "Vẫn còn missing!"
```

## 5. Outlier Analysis
> Phân loại feature trước, xử lý riêng từng nhóm

```python
num_cols  = df.select_dtypes(include=[np.number]).columns
df_num    = df[num_cols]
nunique   = df_num.nunique()
zero_ratio = (df_num == 0).mean()

# Loại flag columns
flag_cols = [c for c in df_num.columns if c.endswith('_was_missing') or c.endswith('_nonzero')]

# Phân loại feature
year_cols     = [c for c in num_cols if ('Year' in c or 'Yr' in c) and c not in flag_cols]
zero_inflated = [c for c in zero_ratio[zero_ratio > 0.3].index if c not in flag_cols and c not in year_cols]
count_cols    = [c for c in nunique[nunique < 10].index if c not in zero_inflated and c not in year_cols and c not in flag_cols]
ordinal_cols_num = [c for c in nunique[(nunique >= 10) & (nunique <= 20)].index
                    if c not in zero_inflated and c not in flag_cols and c not in year_cols]
continuous    = [c for c in set(num_cols) - set(year_cols) - set(count_cols) - set(zero_inflated) - set(ordinal_cols_num)
                 if c not in flag_cols]

print(f"Year/Time     : {len(year_cols)}   → {year_cols}")
print(f"Zero-inflated : {len(zero_inflated)}")
print(f"Count         : {len(count_cols)}")
print(f"Ordinal (num) : {len(ordinal_cols_num)}")
print(f"Continuous    : {len(continuous)}")
```

### 5.1 Outlier Report

```python
def outlier_report(df, cols, feature_type, action):
    """Tạo báo cáo outlier cho một nhóm features."""
    rows = []
    for col in cols:
        series = df[col]
        Q1, Q3  = series.quantile(0.25), series.quantile(0.75)
        IQR     = Q3 - Q1
        lower   = Q1 - 1.5 * IQR
        upper   = Q3 + 1.5 * IQR
        n_out   = ((series < lower) | (series > upper)).sum()
        pct_out = n_out / len(series) * 100

        # Continuous: gợi ý action theo skew
        if feature_type == 'Continuous':
            skew = series.skew()
            if abs(skew) > 1:
                action = 'log transform'
            elif pct_out < 5:
                action = 'capping (IQR)'
            elif pct_out < 15:
                action = 'winsorize'
            else:
                action = 'review distribution'

        rows.append({
            'Column'            : col,
            'Feature Type'      : feature_type,
            'Q1'                : round(Q1, 2),
            'Q3'                : round(Q3, 2),
            'Lower Bound'       : round(lower, 2),
            'Upper Bound'       : round(upper, 2),
            'Outlier Count'     : n_out,
            'Outlier %'         : round(pct_out, 2),
            'Recommended Action': action
        })
    return pd.DataFrame(rows)

# Tổng hợp
full_report = pd.concat([
    outlier_report(df, continuous,      'Continuous',    ''),
    outlier_report(df, zero_inflated,   'Zero-inflated', 'split zero/nonzero + winsorize'),
    outlier_report(df, ordinal_cols_num,'Ordinal',       'treat as ordinal — no IQR'),
    outlier_report(df, count_cols,      'Count',         'no outlier handling'),
    outlier_report(df, year_cols,       'Year/Time',     'logical validation only'),
], ignore_index=True)

full_report
```

### 5.2 Visualize Helper

```python
def plot_hist_box(df, col):
    """Histogram + Boxplot với IQR bounds."""
    data  = df[col]
    Q1, Q3 = data.quantile(0.25), data.quantile(0.75)
    IQR   = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    fig, axes = plt.subplots(1, 2, figsize=(12, 4))

    sns.histplot(data, kde=True, bins=30, ax=axes[0])
    axes[0].axvline(lower, color='red',  linestyle='--', label='Lower bound')
    axes[0].axvline(upper, color='blue', linestyle='--', label='Upper bound')
    axes[0].set_title(f'Histogram — {col}  (skew={data.skew():.2f})')
    axes[0].legend()

    sns.boxplot(x=data, ax=axes[1])
    axes[1].axvline(lower, color='red',  linestyle='--')
    axes[1].axvline(upper, color='blue', linestyle='--')
    axes[1].set_title(f'Boxplot — {col}')

    plt.tight_layout()
    plt.show()
```

### 5.3 Continuous — Winsorize + Log Transform

```python
# Xem phân phối trước khi xử lý
plot_hist_box(df, 'LotArea')
plot_hist_box(df, 'GrLivArea')
```

```python
def winsorize(series, lower_q=0.01, upper_q=0.99):
    """Clip series tại percentile lower và upper."""
    lo = series.quantile(lower_q)
    hi = series.quantile(upper_q)
    return series.clip(lo, hi)

# Các cột không log (có giá trị 0 nhiều → log1p vẫn ok nhưng để riêng)
SPECIAL_NO_LOG = ['TotalBsmtSF']

skew_report = []
for col in continuous:
    raw          = df[col].astype(float)
    skew_before  = raw.skew()
    transformed  = raw.copy()

    # Bước 1: Winsorize nếu skew > 2
    if abs(skew_before) > 2:
        transformed = winsorize(transformed)

    # Bước 2: Log transform nếu vẫn còn skew > 1
    skew_mid = transformed.skew()
    if col not in SPECIAL_NO_LOG and abs(skew_mid) > 1:
        transformed = np.log1p(transformed)

    skew_after = transformed.skew()
    df[col + '_log'] = transformed

    skew_report.append({
        'Column'      : col,
        'Feature Type': 'Continuous',
        'Skew Before' : round(skew_before, 3),
        'Skew After'  : round(skew_after, 3),
        'Action'      : 'winsorize+log' if abs(skew_before) > 2 else ('log' if abs(skew_before) > 1 else 'none')
    })

skew_df = pd.DataFrame(skew_report)
skew_df
```

```python
# Xem phân phối sau khi xử lý
plot_hist_box(df, 'LotArea_log')
plot_hist_box(df, 'GrLivArea_log')
```

### 5.4 Zero-inflated — Winsorize non-zero + Flag

```python
skew_nonzero_report = []
for col in zero_inflated:
    df[col]    = df[col].astype(float)
    zero_mask  = df[col] == 0
    non_zero   = df.loc[~zero_mask, col]

    skew_before = non_zero.skew()

    # Winsorize chỉ trên phần non-zero
    non_zero_win = winsorize(non_zero)
    df.loc[~zero_mask, col] = non_zero_win

    skew_after = non_zero_win.skew()

    # Flag: có giá trị hay không
    df[col + '_nonzero'] = (~zero_mask).astype(int)

    skew_nonzero_report.append({
        'Column'      : col,
        'Feature Type': 'Zero-inflated',
        'Skew Before' : round(skew_before, 3),
        'Skew After'  : round(skew_after, 3),
        'Action'      : 'winsorize non-zero + flag'
    })

skew_nonzero_df = pd.DataFrame(skew_nonzero_report)
skew_nonzero_df
```

### 5.5 Year/Time — Logical Validation

```python
# Clip năm bất thường (< 1800)
for col in year_cols:
    n_anomaly = (df[col] < 1800).sum()
    if n_anomaly > 0:
        print(f"{col}: {n_anomaly} giá trị < 1800 → clip")
    df[col] = df[col].clip(1800, df[col].max())

print("Year/Time validation done.")
```

### 5.6 Post-Processing Report

```python
post_report = pd.concat([skew_df, skew_nonzero_df], ignore_index=True)
print(f"Tổng features đã xử lý outlier: {len(post_report)}")
post_report
```

## 6. Encoding

```python
df_encode = df.copy()

# Báo cáo categorical
cat_cols_enc = df_encode.select_dtypes(include=['object', 'category', 'bool']).columns.tolist()
encode_report = pd.DataFrame([{
    'Column'       : col,
    'Dtype'        : df_encode[col].dtype,
    'Nunique'      : df_encode[col].nunique(),
    'Sample Values': df_encode[col].unique()[:5].tolist()
} for col in cat_cols_enc])

print(f"Categorical columns: {len(cat_cols_enc)}")
encode_report
```

```python
# Ordinal encoding — có thứ tự rõ ràng
quality_map       = {'Ex': 5, 'Gd': 4, 'TA': 3, 'Fa': 2, 'Po': 1}
lotshape_map      = {'Reg': 4, 'IR1': 3, 'IR2': 2, 'IR3': 1}
landslope_map     = {'Gtl': 3, 'Mod': 2, 'Sev': 1}
bsmt_exposure_map = {'Gd': 4, 'Av': 3, 'Mn': 2, 'No': 1}
bsmtfin_map       = {'GLQ': 6, 'ALQ': 5, 'BLQ': 4, 'Rec': 3, 'LwQ': 2, 'Unf': 1}
functional_map    = {'Typ': 7, 'Min1': 6, 'Min2': 5, 'Mod': 4, 'Maj1': 3, 'Maj2': 2, 'Sev': 1}
electrical_map    = {'SBrkr': 5, 'FuseA': 4, 'FuseF': 3, 'FuseP': 2, 'Mix': 1}
paveddrive_map    = {'Y': 3, 'P': 2, 'N': 1}

mapping_dict = {
    'LotShape'    : lotshape_map,
    'LandSlope'   : landslope_map,
    'ExterQual'   : quality_map,
    'ExterCond'   : quality_map,
    'BsmtQual'    : quality_map,
    'BsmtCond'    : quality_map,
    'HeatingQC'   : quality_map,
    'KitchenQual' : quality_map,
    'FireplaceQu' : quality_map,
    'GarageQual'  : quality_map,
    'GarageCond'  : quality_map,
    'BsmtExposure': bsmt_exposure_map,
    'BsmtFinType1': bsmtfin_map,
    'BsmtFinType2': bsmtfin_map,
    'Functional'  : functional_map,
    'Electrical'  : electrical_map,
    'PavedDrive'  : paveddrive_map,
}

for col, mapping in mapping_dict.items():
    if col in df_encode.columns:
        df_encode[col] = df_encode[col].map(mapping).fillna(-1)

print(f"Ordinal encoded: {list(mapping_dict.keys())}")
```

```python
# Nominal encoding — One-Hot
nominal_cols = [
    'MSZoning', 'Street', 'LandContour', 'Utilities', 'LotConfig',
    'Neighborhood', 'Condition1', 'Condition2', 'BldgType', 'HouseStyle',
    'RoofStyle', 'RoofMatl', 'Exterior1st', 'Exterior2nd', 'Foundation',
    'Heating', 'CentralAir', 'GarageType', 'GarageFinish',
    'SaleType', 'SaleCondition'
]
# Chỉ encode những cột còn trong df
nominal_cols = [c for c in nominal_cols if c in df_encode.columns]

df_encode = pd.get_dummies(df_encode, columns=nominal_cols, drop_first=True, dtype=int)
print(f"Shape sau One-Hot: {df_encode.shape}")
print(f"Nominal encoded  : {len(nominal_cols)} columns")
```

```python
# Kiểm tra còn object không
remaining_obj = df_encode.select_dtypes(include='object').columns.tolist()
if remaining_obj:
    print(f"⚠️ Còn object columns: {remaining_obj}")
else:
    print("✅ Không còn object columns")
```

## 7. Feature Engineering

```python
def add_features(df):
    """Thêm các features mới từ features hiện có."""
    # Time-based
    df['HouseAge']        = df['YrSold'] - df['YearBuilt']
    df['YearsSinceRemod'] = df['YrSold'] - df['YearRemodAdd']
    df['Remodeled']       = (df['YearBuilt'] != df['YearRemodAdd']).astype(int)

    # Area-based
    df['TotalArea'] = (
        df['GrLivArea'] + df['TotalBsmtSF'] +
        df['GarageArea'] + df['1stFlrSF'] + df['2ndFlrSF']
    )

    # Missing indicator
    df['HasLotFrontage'] = df['LotFrontage'].notna().astype(int)

    # Log transforms — giảm skew
    df['LogGrLivArea']  = np.log1p(df['GrLivArea'])
    df['LogTotalArea']  = np.log1p(df['TotalArea'])
    df['LogLotArea']    = np.log1p(df['LotArea'])

    # Interaction feature
    df['QualArea'] = df['OverallQual'] * df['GrLivArea']

    return df

df_encode = add_features(df_encode)
print(f"Shape sau Feature Engineering: {df_encode.shape}")
print("New features:", ['HouseAge','YearsSinceRemod','Remodeled','TotalArea',
                        'HasLotFrontage','LogGrLivArea','LogTotalArea','LogLotArea','QualArea'])
```

## 8. Target Analysis — SalePrice

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

sns.histplot(y, kde=True, ax=axes[0], color='steelblue')
axes[0].axvline(y.mean(), color='red', linestyle='--', label=f'Mean: {y.mean():,.0f}')
axes[0].set_title(f'SalePrice — Raw  (skew={y.skew():.3f})')
axes[0].legend()

y_log = np.log1p(y)
sns.histplot(y_log, kde=True, ax=axes[1], color='coral')
axes[1].axvline(y_log.mean(), color='red', linestyle='--', label=f'Mean: {y_log.mean():.2f}')
axes[1].set_title(f'log1p(SalePrice)  (skew={y_log.skew():.3f})')
axes[1].legend()

plt.tight_layout()
plt.show()

print(f"SalePrice skewness     : {y.skew():.3f}  → right-skewed")
print(f"log(SalePrice) skewness: {y_log.skew():.3f}  → gần chuẩn hơn")
print("→ Train model trên log1p(SalePrice) để giảm ảnh hưởng outlier giá cao.")
```

## 9. Final Check & Save

```python
# Ghép X và y
df_clean = pd.concat([
    df_encode.reset_index(drop=True),
    y.reset_index(drop=True)
], axis=1)

print(f"Final shape: {df_clean.shape}")
print(f"Missing    : {df_clean.isnull().sum().sum()}")
print(f"Object cols: {df_clean.select_dtypes(include='object').shape[1]}")
assert df_clean.isnull().sum().sum() == 0, "Còn missing!"
assert df_clean.select_dtypes(include='object').empty, "Còn object!"
print("✅ Clean data ready!")
```

## 10. Summary

| Bước | Xử lý | Kết quả |
|------|--------|---------|
| **Missing** | Drop >50%, flag, fill median/mode | Không còn missing |
| **Outlier — Continuous** | Winsorize (skew>2) + Log (skew>1) | Skew giảm đáng kể |
| **Outlier — Zero-inflated** | Winsorize non-zero + flag `_nonzero` | Giữ thông tin zero |
| **Outlier — Year** | Clip giá trị < 1800 | Loại bỏ năm bất hợp lý |
| **Encoding — Ordinal** | Map theo thứ tự có nghĩa | 17 columns |
| **Encoding — Nominal** | One-Hot (drop_first) | 21 columns → nhiều dummies |
| **Feature Engineering** | Time, Area, Log, Interaction | +9 features mới |
| **Target** | log1p(SalePrice) khi train | Skew: 1.88 → 0.12 |
