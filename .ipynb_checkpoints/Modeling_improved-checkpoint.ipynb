# 🏠 House Price Prediction — Modeling
> Improved version: fixed evaluate(), added Lasso/ElasticNet, proper GridSearchCV

## 1. Library Imports

```python
from pathlib import Path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, KFold, cross_val_score, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.pipeline import Pipeline

try:
    import lightgbm as lgb
    HAS_LGB = True
    print("LightGBM available")
except ImportError:
    HAS_LGB = False
    print("lightgbm not installed — skip LightGBM section")

sns.set_style('whitegrid')
plt.rcParams['figure.dpi'] = 100
RANDOM_STATE = 42
```

## 2. Data Loading

```python
BASE_DIR = Path.cwd().parent
DATA_PATH = BASE_DIR / "Data" / "processed" / "preprocessed.csv"
csv_data = pd.read_csv(DATA_PATH)
print(f"Shape: {csv_data.shape}")
csv_data.head()
```

## 3. Target Analysis & Log Transform

```python
df = csv_data.copy()
X = df.drop("SalePrice", axis=1)
y = df["SalePrice"]

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
sns.histplot(y, kde=True, ax=axes[0])
axes[0].set_title(f'SalePrice  (skew={y.skew():.3f})')

y_log = np.log1p(y)
sns.histplot(y_log, kde=True, ax=axes[1], color='orange')
axes[1].set_title(f'log(SalePrice)  (skew={y_log.skew():.3f})')
plt.tight_layout()
plt.show()

print(f"SalePrice skewness    : {y.skew():.3f}")
print(f"log(SalePrice) skewness: {y_log.skew():.3f}")
print("→ Training models on log(SalePrice)")
```

## 4. Train/Test Split

```python
X_train, X_test, y_train_log, y_test_log = train_test_split(
    X, y_log, test_size=0.2, random_state=RANDOM_STATE
)

# Keep raw y for metric reporting
y_train_raw = np.expm1(y_train_log)
y_test_raw  = np.expm1(y_test_log)

print(f"X_train: {X_train.shape}  |  X_test: {X_test.shape}")
print(f"y_train_raw range: [{y_train_raw.min():.0f}, {y_train_raw.max():.0f}]")
print(f"y_test_raw  range: [{y_test_raw.min():.0f},  {y_test_raw.max():.0f}]")
```

## 5. Preprocessing — StandardScaler
> Fit on train only, transform both train and test (no data leakage)

```python
num_cols  = X_train.select_dtypes(include=[np.number]).columns
flag_cols = [c for c in num_cols if c.endswith('_was_missing') or c.endswith('_nonzero')]
scale_cols = [c for c in num_cols if c not in flag_cols]

print(f"Total numeric cols : {len(num_cols)}")
print(f"Flag cols (skip)   : {len(flag_cols)}")
print(f"Cols to scale      : {len(scale_cols)}")
```

```python
X_train = X_train.copy()
X_test  = X_test.copy()

X_train[scale_cols] = X_train[scale_cols].astype(float)
X_test[scale_cols]  = X_test[scale_cols].astype(float)

scaler = StandardScaler()
X_train.loc[:, scale_cols] = scaler.fit_transform(X_train[scale_cols])
X_test.loc[:, scale_cols]  = scaler.transform(X_test[scale_cols])

# Sanity check
assert X_train.select_dtypes(include='object').empty, "Still has object columns!"
print("Scaling done. Train mean (sample):",
      X_train[scale_cols].mean().abs().max().round(6), "(≈ 0)")
```

## 6. Evaluate Helper Function
> **FIX**: predictions are converted back via `np.expm1()` before computing metrics on raw prices

```python
def evaluate(model, X_tr, X_te, y_tr_raw, y_te_raw, name='Model'):
    """Evaluate a model trained on log(y).
    Predictions are inverse-transformed before computing metrics.
    """
    # Inverse-transform log predictions → real price scale
    y_tr_pred = np.expm1(model.predict(X_tr))
    y_te_pred = np.expm1(model.predict(X_te))

    def _metrics(y_true, y_pred, label):
        r2   = r2_score(y_true, y_pred)
        mae  = mean_absolute_error(y_true, y_pred)
        rmse = np.sqrt(mean_squared_error(y_true, y_pred))
        rmsle = np.sqrt(mean_squared_error(
            np.log1p(y_true),
            np.log1p(np.clip(y_pred, 0, None))
        ))
        print(f"  {label:6s}: R²={r2:.4f}  MAE={mae:>10,.0f}  RMSE={rmse:>10,.0f}  RMSLE={rmsle:.4f}")
        return {'r2': r2, 'mae': mae, 'rmse': rmse, 'rmsle': rmsle}

    print(f"\n{'='*60}")
    print(f"  {name}")
    print(f"{'='*60}")
    tr = _metrics(y_tr_raw, y_tr_pred, 'Train')
    te = _metrics(y_te_raw, y_te_pred, 'Test ')
    gap = tr['r2'] - te['r2']
    if gap > 0.05:
        print(f"  ⚠️  Overfit gap (Train-Test R²): {gap:.4f}")
    else:
        print(f"  ✅ Overfit gap (Train-Test R²): {gap:.4f}")
    return te, y_te_pred


def plot_results(y_true, y_pred, title):
    """Plot Actual vs Predicted and Residual Plot side by side."""
    residuals = y_true - y_pred
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))

    # Actual vs Predicted
    axes[0].scatter(y_true, y_pred, alpha=0.4, s=20)
    lo, hi = y_true.min(), y_true.max()
    axes[0].plot([lo, hi], [lo, hi], 'r--', lw=2)
    axes[0].set_xlabel('Actual Price')
    axes[0].set_ylabel('Predicted Price')
    axes[0].set_title(f'Actual vs Predicted — {title}')

    # Residual
    axes[1].scatter(y_pred, residuals, alpha=0.4, s=20)
    axes[1].axhline(0, color='gray', linestyle='--')
    axes[1].set_xlabel('Predicted Price')
    axes[1].set_ylabel('Residual (Actual − Predicted)')
    axes[1].set_title(f'Residual Plot — {title}')

    plt.tight_layout()
    plt.show()
```

## 7. Baseline — LinearRegression
> With 234 features and only ~1168 samples, OLS is expected to overfit badly (no regularization)

```python
lr = LinearRegression()
lr.fit(X_train, y_train_log)

print(f"Max |coef|   : {np.max(np.abs(lr.coef_)):.4f}")
print(f"Intercept    : {lr.intercept_:.4f}")
print("(Huge coefficients → severe multicollinearity / overfitting)")

lr_metrics, _ = evaluate(lr, X_train, X_test, y_train_raw, y_test_raw, 'LinearRegression')
```

## 8. Ridge (alpha=1.0) — Quick Sanity Check

```python
ridge_base = Ridge(alpha=1.0)
ridge_base.fit(X_train, y_train_log)

print(f"Max |coef|   : {np.max(np.abs(ridge_base.coef_)):.4f}")
print(f"Intercept    : {ridge_base.intercept_:.4f}")

ridge_base_metrics, y_ridge_pred = evaluate(
    ridge_base, X_train, X_test, y_train_raw, y_test_raw, 'Ridge(alpha=1.0)'
)
plot_results(y_test_raw, y_ridge_pred, 'Ridge alpha=1.0')
```

## 9. Cross-Validation Comparison (Ridge / Lasso / ElasticNet)

```python
kf = KFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)

models_cv = {
    'Ridge(1.0)'       : Ridge(alpha=1.0),
    'Lasso(0.001)'     : Lasso(alpha=0.001, max_iter=5000),
    'ElasticNet(0.001)': ElasticNet(alpha=0.001, l1_ratio=0.5, max_iter=5000),
}

print(f"{'Model':<25}  CV R² (mean ± std)")
print('-' * 50)
for name, model in models_cv.items():
    scores = cross_val_score(model, X_train, y_train_log, cv=kf, scoring='r2', n_jobs=-1)
    print(f"{name:<25}  {scores.mean():.4f} ± {scores.std():.4f}")
```

## 10. GridSearchCV — Ridge

```python
alphas = np.logspace(-3, 3, 30)

ridge_gs = Ridge()   # Note: Ridge has no random_state
grid_ridge = GridSearchCV(
    ridge_gs,
    param_grid={'alpha': alphas},
    cv=kf,
    scoring='r2',
    n_jobs=-1,
    verbose=0
)
grid_ridge.fit(X_train, y_train_log)  # FIX: train on log target

print(f"Best alpha : {grid_ridge.best_params_['alpha']:.4f}")
print(f"Best CV R² : {grid_ridge.best_score_:.4f}")
```

```python
# Plot CV score vs alpha
cv_results = pd.DataFrame(grid_ridge.cv_results_)
plt.figure(figsize=(8, 4))
plt.semilogx(cv_results['param_alpha'].astype(float), cv_results['mean_test_score'], marker='o')
plt.axvline(grid_ridge.best_params_['alpha'], color='red', linestyle='--', label=f"Best alpha={grid_ridge.best_params_['alpha']:.3f}")
plt.xlabel('Alpha (log scale)')
plt.ylabel('CV R²')
plt.title('Ridge — GridSearchCV: alpha vs CV R²')
plt.legend()
plt.tight_layout()
plt.show()
```

## 11. GridSearchCV — Lasso

```python
lasso_alphas = np.logspace(-4, 0, 30)

grid_lasso = GridSearchCV(
    Lasso(max_iter=10000),
    param_grid={'alpha': lasso_alphas},
    cv=kf,
    scoring='r2',
    n_jobs=-1
)
grid_lasso.fit(X_train, y_train_log)

print(f"Best alpha : {grid_lasso.best_params_['alpha']:.6f}")
print(f"Best CV R² : {grid_lasso.best_score_:.4f}")

best_lasso = grid_lasso.best_estimator_
n_zero = np.sum(best_lasso.coef_ == 0)
print(f"Features zeroed out: {n_zero} / {len(best_lasso.coef_)} (Lasso feature selection)")
```

## 12. Final Model Comparison

```python
best_ridge = grid_ridge.best_estimator_
best_lasso = grid_lasso.best_estimator_

results = {}
for name, model in [
    ('LinearRegression', lr),
    ('Ridge (best)', best_ridge),
    ('Lasso (best)', best_lasso),
]:
    metrics, _ = evaluate(model, X_train, X_test, y_train_raw, y_test_raw, name)
    results[name] = metrics

comparison_df = pd.DataFrame(results).T.round(4)
print("\n--- Test Set Summary ---")
print(comparison_df.to_string())
```

```python
# Bar chart comparison
fig, axes = plt.subplots(1, 3, figsize=(14, 4))
metrics_to_plot = ['r2', 'mae', 'rmse']
for i, metric in enumerate(metrics_to_plot):
    comparison_df[metric].plot(kind='bar', ax=axes[i], color=['#e74c3c', '#3498db', '#2ecc71'])
    axes[i].set_title(metric.upper())
    axes[i].set_xticklabels(comparison_df.index, rotation=20, ha='right')
    axes[i].set_xlabel('')
plt.suptitle('Test Set — Model Comparison', fontsize=13, y=1.02)
plt.tight_layout()
plt.show()
```

## 13. Best Ridge Model — Detailed Analysis

```python
_, y_best_pred = evaluate(best_ridge, X_train, X_test, y_train_raw, y_test_raw, 'Best Ridge (final)')
plot_results(y_test_raw, y_best_pred, 'Best Ridge')
```

```python
# KDE: Actual vs Predicted
plt.figure(figsize=(8, 5))
sns.kdeplot(y_test_raw, label='Actual', fill=True, color='steelblue', alpha=0.6)
sns.kdeplot(y_best_pred, label='Predicted', fill=True, color='coral', alpha=0.6)
plt.xlabel('House Price')
plt.ylabel('Density')
plt.title('KDE: Actual vs Predicted (Best Ridge)')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Residual distribution
residuals = y_test_raw - y_best_pred
plt.figure(figsize=(7, 4))
sns.histplot(residuals, kde=True, bins=40)
plt.axvline(0, color='red', linestyle='--')
plt.xlabel('Residual (Actual − Predicted)')
plt.title('Residual Distribution (Best Ridge)')
plt.tight_layout()
plt.show()

print(f"Residual mean  : {residuals.mean():,.0f}")
print(f"Residual std   : {residuals.std():,.0f}")
print(f"Residual median: {np.median(residuals):,.0f}")
```

```python
# Top 15 most influential features
coef_series = pd.Series(best_ridge.coef_, index=X_train.columns)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
coef_series.nlargest(15).plot(kind='barh', ax=axes[0], color='steelblue')
axes[0].set_title('Top 15 Positive Coefficients (Best Ridge)')

coef_series.nsmallest(15).plot(kind='barh', ax=axes[1], color='coral')
axes[1].set_title('Top 15 Negative Coefficients (Best Ridge)')

plt.tight_layout()
plt.show()
```
