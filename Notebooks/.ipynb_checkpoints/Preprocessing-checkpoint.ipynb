# 4.Missing Value

```python
# Create a table showing the number of missing data points and the percentage.
miss_count = df.isnull().sum()
miss_percent = miss_count / len(df) * 100
miss_df = pd.DataFrame({'Missing Count': miss_count, 'Missing Percent': miss_percent}).sort_values('Missing Percent', ascending = False)
miss_df[miss_df['Missing Count'] > 0]
```

```python
# Visualize missing percent
top_miss = miss_df[miss_df['Missing Percent'] > 0].head(20)
plt.figure(figsize=(10, 5))
top_miss['Missing Percent'].sort_values().plot(kind='barh', color='salmon')
plt.axvline(50, color='red', linestyle='--', label='50% threshold')
plt.xlabel('Missing %')
plt.title('Top columns with missing values')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#Remove columns where more than 50 percent of the data is missing.
thresh = 50
to_drop = miss_df[miss_df['Missing Percent'] > 50].index.to_list()
df = df.drop(columns = to_drop)
print(f"Dropped {len(to_drop)} columns: {to_drop}")
print(f"Remaining shape: {df.shape}")
```

```python
# Split the data into numbers and object
num_cols = df.select_dtypes(include = [np.number]).columns.to_list()
cat_cols = df.select_dtypes(include = ['object','category','bool']).columns.to_list()
print(f"Numeric cols: {len(num_cols)}")
print(f"Categorical cols: {len(cat_cols)}")
```

```python
# Flag creation is missing
cols_with_flag = df.columns[df.isnull().sum() > 0]
for col in cols_with_flag:
    df[f"{col}_was_missing"] = df[col].isnull().astype(int)
print(f"Flag columns created: {[c for c in df.columns if c.endswith('_was_missing')]}")
```

```python
#Fill in the data using the median by group.
if "Neighborhood" in df.columns:
    df[num_cols] = df.groupby('Neighborhood')[num_cols].transform(lambda x: x.fillna(x.median()))
```

```python
# Fill in the data numeric
df[num_cols] = df[num_cols].fillna(df[num_cols].median())
```

```python
# Fill category using the mode
for col in cat_cols:
    if df[col].isnull().any():
        mode = df[col].mode()
        fill_na = mode.iloc[0] if len(mode) else "Missing"
        df[col] = df[col].fillna(fill_na)

print(f"Missing sau khi fill: {df.isnull().sum().sum()}")
assert df.isnull().sum().sum() == 0, "Vẫn còn missing!"
```

# 5.Ouliers Analysis

```python
num_cols = df.select_dtypes(include = [np.number]).columns
df_num = df[num_cols]
len(num_cols)
```

```python
#Count the number of unique values for each numeric column
nunique = df_num.nunique()
```

```python
#Calculate the proportion of zero values in each numeric column
zero_radio = (df_num == 0).mean()
```

```python
# Flag column filter
flag = [col for col in df_num.columns if col.endswith("_was_missing")]
```

```python
#Year or time-related numeric features
year_condition = num_cols[num_cols.str.contains("Year|Yr", case = False)]
year_feature = [col for col in year_condition if col not in flag]
year_cols = year_feature
print(f"Year/Time: {len(year_cols)}  → {year_cols}")
```

```python
# Numeric features with a high proportion of zero values
zero_condition = zero_radio[zero_radio > 0.3].index
zero_feature = [col for col in zero_condition if col not in flag]
zero_inflated = zero_feature
print(f"Zero-inflated: {len(zero_inflated)}")
```

```python
# Discrete count features (non-negative integers)
count_condition = nunique[nunique < 10].index
count_feature = [col for col in count_condition if col not in zero_feature and col not in year_feature and col not in flag]
count = count_feature
print(f"Count: {len(count)}")
```

```python
# Ordinal features with limited unique values
ordinal_condition = nunique[(nunique >= 10) & (nunique <= 20)].index
ordinal_feature = [col for col in ordinal_condition if col not in zero_feature and col not in flag and col not in year_feature]
ordinal = ordinal_feature
print(f"Ordinal (num): {len(ordinal)}")
```

```python
# Continuous numeric features
continuous_condition = list(set(num_cols) - set(year_cols) - set(count)  - set(zero_inflated) - set(ordinal))
continuous = [col for col in continuous_condition if col not in flag]
print(f"Continuous: {len(continuous)}")
```

## 5.1 Outlier Report

```python
# Create Summary Continuouse
def outlier_report(df, cols, feature_type, action):
    rows = []
    for col in cols:
        series = df[col]
        Q1, Q3 = series.quantile(0.25), series.quantile(0.75)
        IQR    = Q3 - Q1
        lower  = Q1 - 1.5 * IQR
        upper  = Q3 + 1.5 * IQR
        n_out = ((series < lower)|(series > upper))
        out_count = n_out.sum()
        pct_out = out_count / len(series) * 100
    
        # Xác định hành động đề xuất
        if feature_type == 'Continuous':
            skew = series.skew()
            if abs(skew) > 1:
                action = 'log transform'
            elif pct_out < 5:
                action = 'capping (IQR)'
            elif pct_out < 15:
                action = 'winsorize'
            else:
                action = 'review distribution'

        rows.append({
            'Column'            : col,
                'Feature Type'      : feature_type,
                'Q1'                : round(Q1, 2),
                'Q3'                : round(Q3, 2),
                'Lower Bound'       : round(lower, 2),
                'Upper Bound'       : round(upper, 2),
                'Outlier Count'     : out_count,
                'Outlier %'         : round(pct_out, 2),
                'Recommended Action': action
        })
    return pd.DataFrame(rows)

# Full Report
full_report = pd.concat([
    outlier_report(df, continuous, 'Continuous', ''),
    outlier_report(df, zero_inflated, 'Zero-inflated', 'split zero/nonzero + winsorize'),
    outlier_report(df, ordinal,'Ordinal', 'treat as ordinal — no IQR'),
    outlier_report(df, count, 'Count', 'no outlier handling'),
    outlier_report(df, year_cols, 'Year/Time','logical validation only'),
], ignore_index=True)

full_report
```

## 5.2 Visualize Helper

```python
# Plot histogram and boxplot
def plot_hist_box(df, col):
    data = df[col]
    
    q1, q3 = data.quantile(0.25), data.quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr

    fig, axes = plt.subplots(1, 2, figsize = (12,4))
    # Histogram
    sns.histplot(data, kde = True, bins=30, ax = axes[0])
    axes[0].axvline(lower, color = 'red', linestyle="--", label="Lower bound")
    axes[0].axvline(upper, color = 'blue', linestyle="--", label="Upper bound")
    axes[0].set_title(f'Histogram — {col}  (skew={data.skew():.2f})')
    axes[0].legend()

    # Boxplot 
    sns.boxplot(x=data, ax=axes[1])
    axes[1].axvline(lower, color='red',  linestyle='--')
    axes[1].axvline(upper, color='blue', linestyle='--')
    axes[1].set_title(f'Boxplot — {col}')
    
    plt.tight_layout()
    plt.show()
```

## 5.3 Continuous — Winsorize + Log Transform

```python
# Xem phân phối trước khi xử lý
plot_hist_box(df, 'LotArea')
plot_hist_box(df, 'GrLivArea')
```

```python
# Use winsor for skew > 2 and log for skew > 1
def winsorize(data, lower_q = 0.01, upper_q = 0.99):
    lower = data.quantile(lower_q)
    upper = data.quantile(upper_q)
    return data.clip(lower, upper)

# Các cột không log (có giá trị 0 nhiều → log1p vẫn ok nhưng để riêng)
SPECIAL_NO_LOG = ['TotalBsmtSF']

skew_report = []
for col in continuous:
    raw = df[col].astype(float)
    skew_before = raw.skew()
    transformed = raw.copy()
    
    print(f"{col}: {skew_before}")

    # Bước 1: Winsorize nếu skew > 2
    if abs(skew_before) > 2:
        transformed = winsorize(transformed)

   # Bước 2: Log transform nếu vẫn còn skew > 1
    skew_mid = transformed.skew()
    if col not in SPECIAL_NO_LOG and abs(skew_mid) > 1:
        transformed = np.log1p(transformed)

    # Create the log columns
    skew_after = transformed.skew()
    df[col + "_log"] = transformed

    skew_report.append({
        "Columns":col,
        "Feature Type": "Continuous",
        "Skew Before": skew_before,
        "Skew After": skew_after
    })

report_skew_df = pd.DataFrame(skew_report)
report_skew_df
```

```python
plot_hist_box(df, 'LotArea_log')
plot_hist_box(df, 'GrLivArea_log')
```

```python
# Outlier Handling Summary

# Continuous features: log transform or IQR capping

# Zero-inflated features: zero split + flag + winsorization

# Ordinal & count features: no statistical outlier handling

# Year features: logical validation only
```

## 5.4 Zero-inflated - Winsorize non-zero + flag

```python
skew_nonzero_report = []
for col in zero_inflated:
    df[col] = df[col].astype(float)
    zero_mask = df[col] == 0
    non_zero = df.loc[~zero_mask, col]

    skew_before = non_zero.skew()

    # winsorize non-zero only
    non_zero_win = winsorize(non_zero)
    df.loc[~zero_mask, col] = non_zero_win

    skew_after = non_zero_win.skew()

    # Flag: có giá trị 
    df[col + "_nonzero"] = (~zero_mask).astype(int)

    skew_nonzero_report.append({
        "Columns": col,
        "Feature Type": "Non Zero",
        "Skew Before": skew_before,
        "Skew After": skew_after,
        'Action': 'winsorize non-zero + flag'
    })
    
skew_nonzero_report_df = pd.DataFrame(skew_nonzero_report)
skew_nonzero_report_df
```

## 5.5.Year/Time - Logical validation

```python
for col in year_cols:
    n_anomaly = (df[col] < 1800).sum()
    if n_anomaly > 0:
        print(f"{col}: {n_anomaly} giá trị < 1800 → clip")
    df[col] = df[col].clip(1800, df[col].max())

print("Year/Time validation done.")
```

## Post_Processing Report

```python
post_report = pd.concat([report_skew_df, skew_nonzero_report_df], ignore_index=True)
print(f"Tổng features đã xử lý outlier: {len(post_report)}")
post_report
```

```python
# Convert data from string to integer
quality_map = {"Ex": 5, "Gd": 4, "TA": 3, "Fa": 2, "Po": 1}
lotshape_map = {"Reg": 4, "IR1": 3, "IR2": 2, "IR3": 1}
landslope_map = {"Gtl": 3, "Mod": 2, "Sev": 1}
bsmt_exposure_map = {"Gd": 4, "Av": 3, "Mn": 2, "No": 1}
bsmtfin_map = {"GLQ": 6, "ALQ": 5, "BLQ": 4, "Rec": 3, "LwQ": 2, "Unf": 1}
functional_map = {"Typ": 7, "Min1": 6, "Min2": 5, "Mod": 4, "Maj1": 3, "Maj2": 2, "Sev": 1}
electrical_map = {"SBrkr": 5, "FuseA": 4, "FuseF": 3, "FuseP": 2, "Mix": 1}
paveddrive_map = {"Y": 3,"P": 2,"N": 1}
```

# 7.Feature Engineering

```python
def add_features(df):
    # Time-based
    df["HouseAge"] = df["YrSold"] - df["YearBuilt"]
    df["YearsSinceRemod"] = df["YrSold"] - df["YearRemodAdd"]
    df["Remodeled"] = (df["YearBuilt"] != df["YearRemodAdd"]).astype(int)

    # Area-based
    df["TotalArea"] = (df["GrLivArea"] + df["TotalBsmtSF"] + df["GarageArea"] + df["1stFlrSF"] + df["2ndFlrSF"])

    # Missing indicator
    df["HasLotFrontage"] = (~df["LotFrontage"].isna()).astype(int)

    # Log-transforms (giảm skew cho các cột liên tục lớn)
    df["LogGrLivArea"] = np.log1p(df["GrLivArea"])
    df["LogTotalArea"] = np.log1p(df["TotalArea"])
    df["LogLotArea"]   = np.log1p(df["LotArea"])

    # OverallQual
    df["QualArea"] = df["OverallQual"] * df["GrLivArea"]

    return df
```

```python
df_encode = add_features(df_encode)
print(f"Shape sau Feature Engineering: {df_encode.shape}")
```
